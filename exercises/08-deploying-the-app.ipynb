Now our container is ready and pushed to the registry, we can deploy it as a service so end users can access it.

In this notebook, we will deploy our container to Azure Container Apps, both through the portal and from the command line.

## Deploying from the portal

When you're developing, you can quickly test something out by going to the portal and see what's possible.

We will therefore first deploy our container from the portal, after which we'll convert the steps to a single command line command.

> **Exercise** ðŸ“
>
> - Navigate to Azure Container Apps in the Azure portal
> - Click on "Create"
> - Configure the container app
>   - Select your subscription and resource group
>   - Set the container app name to `llmops-app-<your-user-name>`
>   - Select your region (e.g., West Europe)
>   - Create or select a Container Apps Environment
>   - Select your container from the Azure Container Registry
>   - Set the container port to 8081
>   - Enable ingress and set it to accept traffic from anywhere
>   - Set minimum replicas to 0 and maximum to 1
>   - Keep the rest as default
>
> - Once your app is deployed, navigate to the overview and find the application URL
> - Try to access the app from your browser! ðŸŽ‰
> 

## Deploying from the command line

Those were a lot of steps, that don't lend themselves well to automation. 

Let's convert those steps to a single command that we can run from the terminal (and later in CI/CD).

> **Exercise** ðŸ“
>
> - Check out the [docs](https://learn.microsoft.com/en-us/azure/container-apps/containerapp-up) on how to deploy a container app from the command line
> - Use the command below to deploy the service, after filling in the gaps for the image and resource group
> - Once deployed, check the portal if there is indeed a new revision for your container app
> - Try to access the app through the application URL again ðŸŽ‰
> - (Bonus) Add this command to the `makefile`, so we don't have to remember it next time
>
> ðŸ’¡ Hints:
>
> - Check the previous notebook on how we set the environment variables from `.env` in our terminal
> - Check the previous notebook on how we defined the `IMAGE_NAME` with the environment variables

```bash
az containerapp up \
		--name llmops-app-${USER_NAME} \
		--resource-group ${RESOURCE_GROUP} \
		--location westeurope \
		--image $IMAGE_NAME \
		--target-port 8081 \
		--ingress external \
		--registry-username ${CONTAINER_REGISTRY_USERNAME} \
		--registry-password ${CONTAINER_REGISTRY_PASSWORD}

```
