## Push container to registry

Before we can deploy our container in the cloud, the container image needs to be pushed to the cloud.

Prerequisite: make sure an Azure Container Registry exists. If this does not exist yet, ask the trainer to create it.



> **Exercise** ðŸ“
>
> - To push to the container registry, all we need to do is to tag the image with the registry path and push it. This requires some environment variables to be set. Therefore, let's make them available in our terminal:
> 
>    ```bash
>    set -a
>    source .env
>    set +a
>    ```
>
> - Now, define the image name:
>   ```bash
>   IMAGE_NAME=${CONTAINER_REGISTRY}.azurecr.io/llmops-app-${USER_NAME}
>   ```
> - Check if the image name doesn't contain any placeholders still or double quotes:
>   ```bash
>   echo $IMAGE_NAME
>   ```
> - Build and tag the container for the Azure Container Registry:
>   ```bash
>   docker build -t ${IMAGE_NAME} .
>   ```
> - Authenticate Docker to the Azure Container Registry:
>   ```bash
>   az acr login --name ${CONTAINER_REGISTRY}
>   ```
> - Push the container to the Azure Container Registry: 
>   ```bash
>   docker push ${IMAGE_NAME}
>   ```
>
>   ... if you get access denied try running: `TOKEN=$(az acr login --name $CONTAINER_REGISTRY --expose-token --output tsv --query accessToken)` followed by `docker login $CONTAINER_REGISTRY.azurecr.io --username 00000000-0000-0000-0000-000000000000 --password-stdin <<< $TOKEN` as an alternative method of logging in.
>   
> - Check in the Azure portal if the image is available in the container registry. ðŸŽ‰
